# B. Solution

```python
#Drop any duplicate patients in the data
df.drop(df[df['INPATIENT_DATA_ID_x'].duplicated(keep=False)].index, inplace=True)
#Replace data from rows 608 and down from CHAIR to ORDER_STATUS
df.loc[608:, "ORDER_STATUS"] = df.loc[608:, "CHAIR"]
#Replace data from rows 608 and down from Unnamed: 19 to CHAIR
df.loc[608:, "CHAIR"] = df.loc[608:, "Unnamed: 19"]
#Drop the unnecesary column, Unnamed: 19
df = df.drop(columns=['Unnamed: 19'])
```

```python
print(df.loc[608:, [ 'CHAIR', 'ORDER_STATUS', 'APPT_DTTM']])
```

# C. Calculating Wait Times

```python
#Time Differences in Patients Seated After Checking in
START_LATE = df.loc[df['CHAIR_START'].dt.time >= df['CHECKIN_DTTM'].dt.time]
df.loc[START_LATE.index, 'TIME_DIFF (min)'] = (df['CHAIR_START'] - df['CHECKIN_DTTM']).dt.seconds // 60

#Time Differences in Patients Checking Out After being Infused
CHECKOUT_LATE= df.loc[df['CHECKOUT_DTTM'].dt.time >= df['INFUSION_END'].dt.time]
df.loc[CHECKOUT_LATE.index, 'CHECKOUT_INFUSE_DIFF (min)'] = (df['CHECKOUT_DTTM'] - df['INFUSION_END']).dt.seconds // 60

#Time Differences in Patients Infused After being Seated
INFUSE_LATE = df.loc[df['CHAIR_START'].dt.time <= df['INFUSION_START'].dt.time]
df.loc[INFUSE_LATE.index, 'CHAIR_INFUSE_DIFF (min)'] = (df['INFUSION_START'] - df['CHAIR_START']).dt.seconds // 60
```

```python
mean_checkinbefore = df['TIME_DIFF (min)'].mean()
print("Average Wait of Checking in and then getting Seated: ", f'{mean_checkinbefore:.0f}', "minutes")

mean_checkinbefore = df['CHAIR_INFUSE_DIFF (min)'].mean()
print("Average Wait of Seating in and infusing starting: ", f'{mean_checkinbefore:.0f}', "minutes")

mean_checkinbefore = df['CHAIR_INFUSE_DIFF (min)'].mean()
print("Average Wait of Infusing and then checking out: ", f'{mean_checkinbefore:.0f}', "minutes")
```
