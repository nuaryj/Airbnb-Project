# Entity Merge

```python
%%bigquery
select * from books_raw.faker_authors
order by author_id
limit 5
```

```python
%%bigquery
select * from books_raw.author
limit 5
```

```python
%%bigquery
select (select count(*) from books_raw.faker_authors
        where author_id is not null) as faker_count,
        (select count(*) from books_raw.author) as bird_count
```

# author_id

```python
%%bigquery
  create or replace table books_stg.Authors as
    select b.author_id as author_id,
    b.author_name, f.date_of_birth, f.age, f.country, f.email, 'bird_faker' as data_source,b.load_time
    from books_raw.author b
    join books_raw.faker_authors f
    on b.author_id = f.author_id;
```

```python
%%bigquery
  select * from books_stg.Authors
  where author_id is not null
  limit 5
```

```python
%%bigquery
  select * from books_stg.Authors
  where author_id is not null
  limit 5
```

```python
%%bigquery
  select count(*) author_count
  from books_stg.Authors
  where author_id is not null
```

```python
%%bigquery
  select (select count(*) from books_stg.Authors) as merged_table_count,
  (select count(*) from books_raw.author) as bird_table_count
```

# Primary Key

```python
%%bigquery
alter table books_stg.Authors
  add primary key (author_id) not enforced;
```

```python
%%bigquery
select author_id, count(*) duplicate_records
from books_stg.Authors
group by author_id
having count(*) > 1
order by count(*) desc
```
